## **What is protein aggregation?**

Protein aggregation is what occurs when misfolded proteins - full or partial conformational dissonance - clump together with other healthy proteins, forming disordered or ordered assemblies that cells typically cannot use and almost certainly cannot get rid of. There are generally considered to be two types of aggregates: amorphous aggregates & amyloid fibrils. Amorphous aggregates are basically just disorganized clumps, amyloid fibrils are more relevant as they are highly ordered and form cross-β sheet structures. This ordering and cross-β sheet structures makes them thermodynamically stable, highly resistant to proteases, and tend to propagate by templating futher misfolding. 

### *How does this happen?*

Cells typically have "quality control" parameters to prevent this, for example:
- Molecular chaperones: His70, His90, etc. ensure proper folding and fixing conformationally stressed proteins
- Ubiquitin-Proteosome System (UPS): Tags & degrades misfolded proteins
- Autophagy: Cells automatic system for recycling damaged components (like misfolded proteins) 

Aggregation only occurs when all of these systems are overwhelmed, either by genetic mutation or simply sheer concentrations of an aggregation-prone sequence.

### *Protein Aggregation in Human Diseases*

Protein aggregation is an identifying feature of a whole class of diseases called *proteinopathies*. Some examples are:
- Alzheimers: Aggregation of amyloid-β (Aβ) into plaques extraceullularly, and tau (a protein that's supposed to stabilize the internatl structure of neurons) in neurofibrillary tangles intraceulluarly
- Parkinsons: α-synuclein aggregates into Lewy Bodies (abnormal clumps) in dopeminergic neurons
- ALS/FTD:  TDP-43 and FUS mislocalizes from the nucleus and form into cytoplasmic inclusions
- Huntingtons: polyglutamine (polyQ) expansion in huntingtin drives aggregation; longer repeats = earlier onset

### *How are they toxic?*

This is where stuff gets complicated, because the mechanism in which aggregation occurs still isn't really known. The leading hypothesis is that oligomers (100> monomer units) are the toxic species, as opposed to the fibrils themselves. Oligomers can pass through membranes easier, disrupt synaptic functions, and trigger inflammation. Some suspect that the large fibrous units might actually be protective in a sense by sequestering away the more dangerous soluble forms.
