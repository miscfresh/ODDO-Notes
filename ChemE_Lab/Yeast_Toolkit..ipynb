# ***A Highly Characterized Yeast Toolkit for Modular, Multipart Assembly***
*Journal Review*

## **Abstract**
*Saccharomyces cerevisiae* or Brewer's Yeast, is an interesting host for synthetic biology because of it's historical use in industrial fermentations. Research has typically orbited around the use of bacteria, but yeast provides interesting advantages that should be considered. This article presents a novel & versatile engineering platform for yeast that contains a rapid, modular assembly method and a basic set of characterized parts.
