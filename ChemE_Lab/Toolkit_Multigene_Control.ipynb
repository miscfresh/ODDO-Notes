# ***A Toolkit for Precise, Multigene Control in Saccharomyces cerevisiae***
*Journal Review - https://pubs.acs.org/doi/10.1021/acssynbio.2c00423*

## **Abstract**
Researchers rely on systems that allow us to precisely control gene expression. Unfortunately,<a id='fn1'></a>[very few inducible gene expression systems exist that gives us simultaneous multigene control under nutritionally favorable conditions](#fn1_def) in the important model organism and chassis *Saccharomyces cerevisiae* (yeast). The authors of this paper repurposed <a id='term_1'></a>[ligand biniding domains from mammalian type 1 nuclear receptors](#def_1) to establish 5 different types of gene expression systems.
***
By using mammalian LBDs, these researchers have essentially created a system to control gene expression in yeast. Remember, the main function of type 1 nuclear receptors are to receive a ligand, undergo a conformational change, move to the nucleus, and (the receptor -> transcription factor) binds to the DNA sequence where the promotor is located. So if we made a gene with an aldosterone promotor (pAld) and introduced aldosterone (ald) to the yeast cells, the ald would bind to the LBD, which would move into the nucleus and express the gene. It's *gene*-ius!

## **Footnotes**
<a id='fn1_def'></a>[Footnote 1:](#fn1) There aren't many tools that let you turn gene expression on/off in a controlled, switchable way. Specifically, the currents systems that *can* do this can only control one gene at a time. Tuning multiple genes independently and simultaneously is the harder/rarer thing thats implied. Regarding the "nutritonally favorable" bit, most inducible systems that do exists require you to add things like galactose or deprive cells of a nutrient to trigger induction. These conditions can muddy your experiment because of the stress on the cells/change in their metabolic state; it's possible what you end up observing could be in response to that.

<a id='def_1'></a>[Nuclear Receptors:](#term_1) A class of proteins that, when bound to a specific small molecule (the ligand), change their behavior. Typically they change by moving into the nucleus and turning genes on or off. Type 1 receptors normally sit in the cytoplasm and only travel once bound to their ligand (hormone receptors, etc.). The mention of mammalian type 1 receptors is important because yeast don't normally have these receptors. They're saying by introducing a mammalian LBD into a yeast cell, it still responds to its ligand
