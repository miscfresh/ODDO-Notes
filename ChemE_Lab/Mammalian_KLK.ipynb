## **KLK2**

### *Structure-Function Analyses of Human Kallikrein-related Peptidase 2 Establish the 99-Loop as Master Regulator of Activity*
*https://pmc.ncbi.nlm.nih.gov/articles/PMC4256358/*
- Ile-16 -- Asp-194 salt bridge is vital for KLK2 function post-junction cleavage; inactive KLK structures are the ones missing the activating salt bridge between the new N-terminus and Asp-194.
- In this article they actually engineered a self activating variant, so not super applicable to our needs regarding autoactivation
- Did find some unique things about KLK2:
    - Reported crystal structures (1.9 Å) of two KLK2-small molecule inhibitor complexes in which discontinuous electron density for the 99-loop indicates the loop is largely disordered.
    - Glycan-gated activity (dunno)
    - Zn<sup>2+</sup> located at the 99-loop/active site interface. Elimination of this region prevented irreversible inactivation => KLK2 both autoactivated and autolytically destroys itself
- great substrate specificity data in this article. skala et al. ultimately determined the substrate regions P1-P4 to trend like so:
    - P1 => Arg (lys showed some activity too but arg preferred ~2 fold)
    - P2 => aromatic; Phe/Try/Trp
    - P3 => charged, polar, and small side chains; aa like glu, ala(?), gln, ser, etc. 
    - P4 => hyrophobic and to a lesser extent polar; trp strongly preferred over pro, met, etc.
    - note that substrate specificity on position P3-P4 are less pronounced
- they were also able to successfully purify the KLK2 protein (benzamidine-inhibited; KLK2-BEN) and determine it's crystal structure via PISA, confirmed by size exclusion chromatography and dynamic light scattering.
- KLK2 doesn't have one specific target, but more so a list of possible targets:
    - semenogelin I and II
    - fibronection
    - pro-KLK3 region
    - IGFBPs 2-5
    - cancer-context targets like pro-uPA, PAI-1, and PAR-2

### *Activation of Latent Protease Function of Pro-HK2,But Not Pro-PSA, Involves Autoprocessing*
*https://onlinelibrary.wiley.com/doi/10.1002/pros.1088*
- authors found that hk2 (KLK2) could hydrolyze pro-hk2 substrate suggesting that maturation of pro-hk2 to hk2 involves self-activation (autoprocessing); reliant on there already being hk2 present (pro-forms can still react *very* slowly with one another to generate the active hk2)
- pro-hk2 has the sequence VPLIQSR; echoes the previous paper stating incorrect processing of the pro-region affects the folding of the N-terminal portion of the protein (salt bridge)
- presence of Arg at the cleavage site suggest trypsin-like substrate specificity
- old article, doesnt tell us much regarding substrate specificity or protein structure (besides sequence)

### Quick Notes

P*n* - AA or P*n*': P*n* is the amino acid on the left (N-terminal) of the scissile bond, P*n*' is the amino acid on the right (C-terminal). The paper uses the Schechter-Berger system to describe the locations of interaction between the enzymes and substrate. The matching pockets on the enzyme (S2 S1 ---- S1' S2') hold their substrate counterparts. So "P1 - Arg" just means the substrate of KLK2 has an arginine sitting in the P1 position.

Tryptic - "Trypsin-like"; cuts after basic residues like Arg/Lys

Chymotryptic - cuts after bulky aromatics like Phe/Tyr/Trp

**KLK2 is tryptic while KLK3 is chymotryptic**

Junction - the specific peptide bond between the propeptide and the mature enzyme body.

EK (Enterokinase) - useful tool. its a protease thats very strict with its substrate, *it only cuts after DDDDK*. authors of the skala article exploited this by building an artifical peptide ending in DDDDK so they could add EK externally and have it snip precisely at the junction to activate KLK2 on demand. **DESIGNED FOR ACTIVATION, NOT SELF-ACTIVATION!**

## **KLK5**
### *Biochemical and Enzymatic Characterization of Human Kallikrein 5 (hK5), a Novel Serine Protease Potentially Involved in Cancer Progression*
*https://www.sciencedirect.com/science/article/pii/S0021925820659816?via%3Dihub*
- huge paper, studied KLK5 with a yeast expression system!
- took the gene for pro-KLK5 and inserted it into yeast plasmid, attaching an "export tag" at the (lined up/in frame). when they put the plasmid into yeast, the cells built the pro-KLK5 and secreted it into the broth where it was able to be collected and self-activate
- secretion of KLK5 into the yeast supernatant was confirmed via SDS-PAGE; purification was possible (two step method).
- KLK5 displayed trypsin-like activity; KLK5 exhibits much higher preference for Arg at P1 relatuve to Lys (another theorized aa at P1).
- best substrate for KLK5 was determined to be Boc-Val-Pro-Arg-AMC
- "the full-length cDNA encoding pro-KLK5 was amplified via reverse transcriptase PCR using RNA isolated from a breast cell strain, then cloned into the pPIC9 vector in-frame with the yeast alpha-mating factor, and successfully transformed into KM71 strain of the yeast, *P. pastoris*"
- pH of 8 was optimal for KLK5 function with it being 2.5 times more active in phosphate buffer

### *A Proteolytic Cascade of Kallikreins in the Stratum Corneum*
*https://pubmed.ncbi.nlm.nih.gov/15654974/*
- another huge paper. clear evidence of auto-activation of pro-KLK5 -> KLK5! 
- auto-activation of KLK5 was found by incubating purified pro-KLK5 at 37°C, pH of 8, with substrate S-2302 (H-D-Pro-Phe-Arg-pNA + 2HCl). KLK5's activity toward the substrate was measured at different time points and they found there was a time-dependent increase in catalytic activity. Once activity was plotted, a sigmoid curve was obtained (indicating cooperativity) which indicates auto-activation.
    - note the convention in which the substrate was written. H tells us that the N-terminus of the peptide is free/unblocked (H-N); a plain NH2 (no protecting group). D is the stereochemical tag, its the D-enantiomer of proline. this prevents exopeptidases from chewing the free N-terminus (exopeptidases only recognize L-residues). pNA is the C-terminal reporter; fluorescent tag basically. 2HCl is the counterion/salt.
    - kinetics can be explained the the slow initial lag while trace active enzymes begins activation -> exponential activation -> substrate-binding cooperativity (?)
- data shows evidence of a auto-activation cascade across kallikreins 5, 7, and 14. basically as one auto-activates, it assists in activating across the others.
- purification of KLK5 was done from confiditoned cell medium by batch adsorption to CM Sepharose FF, followed by reverse phase chromatography.
- for a cascade experiment:
    - KLK5 could activate pro-KLK7 (pH 5-7), pro-KLK14, and pro-KLK5
    - KLK14 could activate pro-KLK5 but NOT pro-KLK14 (no auto-activation)
    - KLK14 could NOT activate pro-KLK7 under any conditions tested; KLK5 could activate pro-KLK7 at a slow but significant rate.
        - confirmed removal of propeptide region of KLK7 via immunoblotting/western blotting & zymography
    - KLK7 activation was followed by measuring activity against substrate S-2586
        - S-2586:  MeO - Suc - Arg - Pro - Tyr - pNA + HCl (counterion)
        - MeO: methoxy; Suc: succinyl
    - KLK14 actually had *some* chymotrypic activity; could be a way to differentiate between KLK5 if we do a cascade
    - KLK14 doesn't show much promise to be self-activating but is the "better" protease when compared to KLK5; the smallest difference in k<sub>cat</sub> was 10-fold (S-2302), and the highest was 80-fold with S-2288.

### *Human Tissue Kallikrein 5 Is a Member of a Proteolytic Cascade Pathway Involved in Seminal Clot Liquefaction and Potentially in Prostate Cancer Progression*
*https://www.jbc.org/article/S0021-9258(19)46607-6/fulltext*
- provides clear evidence that KLK5 is involved in a cascade involving KLK2 and KLK3!
- they proved that KLK5 was able to activate pro-KLK2 and 3, as well as internally cleave and deactivate them
- Zn<sup>2+</sub> *can* efficiently inhibit KLK5 activity by 97.5%
- KLK5 was able to cleave the inserted heptapeptides encompassing the cleavage sites for KLK1, 2, and 3 with high efficiency; KLK5, 9, 11, and 12 with moderate efficiency; KLK7, 8 and 15 with low efficiency. No cleavage was observed for the other kallikreins.
- after incubation of pro-KLK2 and pro-KLK3 with KL5, activation was demonstrated (verified by western blot)
- internal cleavage appears to be a common method of kallikrein regulation *in vivo*

### Quick Notes
cooperativity - the observation that binding of the sisbtrate or ligand at one binding site affects the affinity of other sites. creates sigmoidal curves when plotting enzymatic kinetics (tell-tale sign). implies allosteric binding

reverse phase chromatography (RP-LC) - a liquid chromatography technique that seperates compounds based on their hydrophobicity. the standard for polar and biological molecules.

immunoblotting - its literally a western blot. highly sensitive method for identification of proteins.

zymography - an electrophoretic technieque used to detect and characterize proteolytic enzymes

*https://www.sciencedirect.com/science/article/pii/S002192582043008X*
- paper contains details on the relationships between a variety of kallikreins, looking into the autolytic and cross-activation relationships.
- KLK3 (PSA) is a widely used prostate cancer biomarker, recent studies point to the ratio of [KLK2]:[KLK3] is an important cancer diagnostic tool
    - note: possible cross-activation between KLK2 and KLK3; KLK3 is chymotrypic and the *Denmeade* paper mentioned earlier stated KLK3 was unable to hydrolyze both pro-KLK2 and pro-KLK3 so it likely cannot auto-activate

- the rest of the kallikreins (3, 7, 9, and 15) are expected to be chymotryptic
