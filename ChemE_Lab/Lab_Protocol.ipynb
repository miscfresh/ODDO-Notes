## ***Miniprep***
*"We do these all the time, you'll pick it up quickly" - Lawton*

If I learn anything from the lab at all, it'll be how to do a miniprep. Minipreps are a method of DNA extraction and purification for <a id='ref-plas'></a>[plasmid DNA](#def-plas). Minipreps are done on inoculated - liquid bacterial cultures, that have been <a id='ref-trans'></a>[transformed](#sec-trans) and isolated. The first step in a miniprep is <a id='ref-lys'></a>[lysis](#def-lys) of the bacteria. There are many methods to lyse a cell: alkaline, mechanical, enzymatic, etc. but in the lab we use an alkaline method. The first step is to transfer the liquid bacteria culture from the culture tube to a mini centrifuge capsule and add a basic solution like sodium hydroxide (NaOH) to disintegrate the cell wall. Once the cell is lysed, we have to then clean up the byproducts and isolate the plasmid of interest. When bacteria is lysed at a high pH (NaOH), the proteins and chromosomal DNA are denatured, but the plasmids remain stable. What's fascinating is that the plasmid DNA also denatures, but because the two strands are interwoven (two overlapping rings) they can easily and accurately refold once exposed to potassiun acetate.

## ***Glossary***
**Plasmid DNA**: A <a id='def-plas'></a>[plasmid](#ref-plas) is a small circular DNA molecule found in bacteria and some other microscopic organisms. Plasmids are physically seperated from chromosomal DNA and replicate independetly. They contain a small number of genes. Ones often attributed to genes that're involved in antibiotic resistance.

**Lysis**: <a id='def-lys'></a>[Lysis](#ref-lys) is the breakdown or disintegrataion of a cell. It is the process of rupturing cell membranes to extract their internal contents for scientific study.
