```python
from rdkit import Chem
from rdkit.Chem import AllChem, Descriptors
from rdkit.Chem.Draw import SimilarityMaps

from rdkit.Chem.Draw import IPythonConsole
IPythonConsole.ipython_useSVG = True

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#morphine = Chem.MolFromSmiles('CCN1CCC23C4C1CC5=C2C(=C(C=C5)O)OC3C(C=C4)O')
#morphine - kind of looks like shit

morphine = Chem.MolFromInchi('InChI=1S/C17H19NO3/c1-18-7-6-17-10-3-5-13(20)16(17)21-15-12(19)4-2-9(14(15)17)8-11(10)18/h2-5,10-11,13,16,19-20H,6-8H2,1H3')
Chem.Draw.MolToImage(morphine, size=(300,300))
```

```python
# download image to SVG

Chem.Draw.MolToFile(morphine, 'morphine.svg',
                      size=(300,300),
                      imageType='svg')
```

```python
# find chiral centers i guess
Chem.FindMolChiralCenters(morphine)
```
