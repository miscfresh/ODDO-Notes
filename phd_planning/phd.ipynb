# **Roadmap for PhD**

## **Schools of Interest**

### *University of Michigan - Biological Chemistry* https://medschool.umich.edu/programs-admissions/phd-programs/pibs-phd-programs/phd-program-biological-chemistry

Located in Ann Arbor, the programs primary research areas are:
- Regulation of gene expression
- **Protein folding and processing**
- **Macromolecular structure and mechanism**
- Biochemical signaling

The biological chemistry PhD program is part of the medical school and is not an easy program to get into. Technically they have a 18% acceptance rate but the average applicant had an undergrad GPA is a 3.73, 3 years of prior research experience, and published abstracts/papers. It'll be an uphill battle but some other programs at U Mich. may be worth looking into.

Requirements for application:
- Select two Ph.D. program/department choices in rank order
- Provide a statement of purpose
- Submit a personal statement
- Solicit 3 letters of recommendation
- Upload scanned copy of official transcripts (if cannot be physical, get etranscript, print, and scan that)
- **Submit by December 1st**

### *University of Illinois Urbana-Champaign - Biochemistry* https://catalog.illinois.edu/graduate/las/biochemistry-phd/

Located in Urbana-Champaign, the programs primary research areas are:
- Physical approaches to the structure and function of macromolecules and membranes
- **Nucleic acid biochemistry and enzymology**
- **Enzyme mechanisms and evolution** (legit combo or Dr. Stewart and Dr. Denard)
- Membrane biochemistry and bioenergetics
- Protein-lipid interactions
- Protein-nucleic acid interactions and molecular recognition
- Molecular biological approaches to gene organization and expression
- Immunology
- Microbial physiology and signal transduction

The biochemistry PhD program here is pretty cracked. I didn't know this but apparently UIUC is a top public school in the US. Admission requirements seem *marginally* easier than U Mich but not by much. They do explicitly state "applicants must have a strong background in chemistry, biology, *physics*, and calculus." which is kind of a red flag because I did pretty cheeks in physics here. Chem and bio are obviously perfectly chill but I dunno how I could prove proficiency in calculus besides my grades at FIU (maybe tutoring? idk). 

Requirements for application:
- GPA >= 3.0
- Prior research experience
- 3 letters of recommendation - Letters must attest to applicants academic and research background
- **Submit by December 1st**

### *University of California San Diego - Biochemistry & Biophysics* https://www-chem.ucsd.edu/graduate-program/bmb/bmb-admissions/index.html

Located in, unsuprisingly, San Diego, the programs primary research areas are (there are a FUCK ton so I'm gonna cut some out):
- Natural product synthesis/biosynth, enzymology, and metabolic engineering
- Neurodegenerative disease, posttranslational modifications
- Molecular mechanism of protein synthesis
- Protein stucture, dynamics and folding
- *and much much more*

Kind of following a downward progression in (perceived) intensity of the colleges. UC San Diego seems exactly as a California college would be: very relaxed but to the point. Not diminishing their work at all, but just the page feels more tranquil. Regardless, this program has a greater emphasis on the biophysics of biochemistry which should pair well with what I want to research and study. While I'm fairly unfamiliar with biophysics as a whole, I'm sure I'll do well in it. Their acceptance rate is higher than UOM's at 35%, but also has significantly higher amount of applicants. Their admissions process seems to be more holistic, which probably explains the higher acceptance rate.

Requirements for application **Verify as application date nears**:
- Scan & upload copies of unofficial transcripts for all institutions attended after high school
- Statement of purpose - answer the following 5 prompts: personal/professional motivations, previous research experience, what you learned, specify 1+ faculty members to work under, and describe obstacles faced throughout your life
- 3 letters of recommendation that provide context and clarity with respect to my ability and potential to excel in a graduate program
- Upload PDFs of research products (papers, posters, abstracts, etc.)
- Additional education experiences (*Optional*)
- **Submit by December 3rd**
